```python
import os
from PIL import Image

def convert_jpg_to_webp(folder_path):
    # Check if the folder exists
    if not os.path.exists(folder_path):
        print(f"The folder {folder_path} does not exist.")
        return

    # Loop through all files in the folder
    for filename in os.listdir(folder_path):
        # Check if the file is a JPG image
        if filename.endswith(".jpg"):
            # Construct full file path
            file_path = os.path.join(folder_path, filename)
            # Open the image
            with Image.open(file_path) as img:
                # Define the new filename with .webp extension
                new_filename = filename[:-4] + ".webp"
                new_file_path = os.path.join(folder_path, new_filename)
                # Convert and save the image as WEBP
                img.save(new_file_path, "WEBP")
                print(f"Converted and saved {new_filename}")

# Example usage
folder_path = '/Users/abdul/Documents/image'
convert_jpg_to_webp(folder_path)
```

```python
import os

def remove_jpg_files(folder_path):
    # Check if the folder exists
    if not os.path.exists(folder_path):
        print(f"The folder {folder_path} does not exist.")
        return

    # Loop through all files in the folder
    for filename in os.listdir(folder_path):
        # Check if the file is a JPG image
        if filename.endswith(".jpg"):
            # Construct full file path
            file_path = os.path.join(folder_path, filename)
            # Remove the JPG file
            os.remove(file_path)
            print(f"Removed {filename}")

# Example usage
remove_jpg_files(folder_path)
```
